```python
def similarity_score(s1, s2):
    value = editdistance.eval(s1, s2)
    return 1 - value / max(len(s1), len(s2))
```

```python
import numpy as np

# Assuming 'similarity_score' is a defined function that calculates the similarity score
# between two descriptions.

# Ensure df and df1 are correctly sized or handle cases where they are not:
num_rows_df = len(df)
num_cols_df1 = min(len(df1.columns), num_rows_df)  # Assuming each row of df1 has same number of columns

# Initialize 'final' as a list of empty lists to store similarity scores in a 2D structure
final = [[] for _ in range(num_rows_df)]

for i in range(num_rows_df):
    for j in range(num_cols_df1):
        # Calculate similarity score and store it directly in 'final'
        score = similarity_score(df["Setting and Description"][i], df1.iloc[i][j])
        final[i].append(score)
        #print(score)

# If you need 'final' to be a numpy array instead of a list of lists:
final_array = np.array(final)
#shape of final array

final_array
```
